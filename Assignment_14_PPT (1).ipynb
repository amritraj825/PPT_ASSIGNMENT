Given a linked list ofÂ **N**Â nodes such that it may contain a loop.

A loop here means that the last node of the link list is connected to the node at position X(1-based index). If the link list does not have any loop, X=0.

Remove the loop from the linked list, if it is present, i.e. unlink the last node which is forming the loop.

**Example 1:**
Input:
N = 3
value[] = {1,3,4}
X = 2
Output:1
Explanation:The link list looks like
1 -> 3 -> 4
     ^    |
     |____|
A loop is present. If you remove it
successfully, the answer will be 1.
**Example 2:**

```
Input:
N = 4
value[] = {1,8,3,4}
X = 0
Output:1
Explanation:TheÂ Linked list does not
contains any loop.
```

**Example 3:**

````python
class ListNode:
    def __init__(self, val):
        self.val = val
        self.next = None


def removeLoop(head):
    if head is None or head.next is None:
        return head

    slowPtr = head
    fastPtr = head

    # Check if loop exists
    while fastPtr and fastPtr.next:
        slowPtr = slowPtr.next
        fastPtr = fastPtr.next.next
        if slowPtr == fastPtr:
            break

    # No loop found
    if slowPtr != fastPtr:
        return head

    # Move slowPtr to head and find loop start
    slowPtr = head
    while slowPtr.next != fastPtr.next:
        slowPtr = slowPtr.next
        fastPtr = fastPtr.next

    # Remove the loop
    fastPtr.next = None

    return head


# Create a linked list from a list of values
def createLinkedList(values):
    if not values:
        return None

    head = ListNode(values[0])
    curr = head
    for val in values[1:]:
        curr.next = ListNode(val)
        curr = curr.next

    return head


# Print the values in a linked list
def printLinkedList(head):
    values = []
    curr = head
    while curr:
        values.append(curr.val)
        curr = curr.next
    print(' -> '.join(str(val) for val in values))


# Test the code with example inputs
values = [1, 3, 4]
head = createLinkedList(values)
print('Input Linked List:')
printLinkedList(head)
head = removeLoop(head)
print('\nLinked List after removing loop:')
printLinkedList(head)
````

A numberÂ **N**Â is represented in Linked List such that each digit corresponds to a node in linked list. You need to add 1 to it.

**Example 1:**

```
Input:
LinkedList: 4->5->6
Output:457

```

**Example 2:**

</aside>

````python
class ListNode:
    def __init__(self, val=0):
        self.val = val
        self.next = None


def addOne(head):
    # Reverse the linked list
    prev = None
    curr = head
    while curr:
        next_node = curr.next
        curr.next = prev
        prev = curr
        curr = next_node
    head = prev

    # Add 1 to the number
    curr = head
    carry = 1
    while curr:
        sum = curr.val + carry
        curr.val = sum % 10
        carry = sum // 10
        if carry == 0:
            break
        curr = curr.next

    # Handle carry if it exists
    if carry > 0:
        new_node = ListNode(carry)
        curr.next = new_node

    # Reverse the linked list again
    prev = None
    curr = head
    while curr:
        next_node = curr.next
        curr.next = prev
        prev = curr
        curr = next_node
    head = prev

    return head


# Create a linked list from a list of digits
def createLinkedList(digits):
    if not digits:
        return None

    head = ListNode(digits[0])
    curr = head
    for digit in digits[1:]:
        curr.next = ListNode(digit)
        curr = curr.next

    return head


# Convert a linked list to a list of digits
def linkedListToDigits(head):
    digits = []
    curr = head
    while curr:
        digits.append(curr.val)
        curr = curr.next

    return digits


# Test the code with example inputs
digits = [4, 5, 6]
head = createLinkedList(digits)
print('Input Linked List:')
print(linkedListToDigits(head))
head = addOne(head)
print('\nLinked List after adding 1:')
print(linkedListToDigits(head))
````

Given a Linked List of size N, where every node represents a sub-linked-list and contains two pointers:(i) aÂ **next**Â pointer to the next node,(ii) aÂ **bottom**Â pointerÂ to a linked list where this node is head.Each of theÂ sub-linked-list is in sorted order.Flatten the Link List such that all the nodes appear in a single level while maintaining the sorted order.Â **Note:**Â The flattened list will be printed using the bottom pointer instead of next pointer.

**Example 1:**

```
Input:
5 -> 10 -> 19 -> 28
|     |     |     |
7     20    22   35
|           |     |
8          50    40
|                 |
30               45
Output:Â 5-> 7-> 8- > 10 -> 19-> 20->
22-> 28-> 30-> 35-> 40-> 45-> 50.
Explanation:
The resultant linked lists has every
node in a single level.(Note:| represents the bottom pointer.)

```

**Example 2:**

</aside>

````python
class ListNode:
    def __init__(self, val=0, next=None, bottom=None):
        self.val = val
        self.next = next
        self.bottom = bottom


def mergeLists(list1, list2):
    if list1 is None:
        return list2
    if list2 is None:
        return list1

    merged = None
    if list1.val <= list2.val:
        merged = list1
        merged.bottom = mergeLists(list1.bottom, list2)
    else:
        merged = list2
        merged.bottom = mergeLists(list1, list2.bottom)

    return merged


def flattenLinkedList(head):
    if head is None or head.next is None:
        return head

    head.next = flattenLinkedList(head.next)
    head = mergeLists(head, head.next)

    return head


# Print the flattened linked list
def printFlattenedList(head):
    if head is None:
        return

    curr = head
    while curr:
        print(curr.val, end=" ")
        curr = curr.bottom
    print()


# Test the code with example inputs
head = ListNode(5)
head.next = ListNode(10)
head.next.next = ListNode(19)
head.next.next.next = ListNode(28)

head.bottom = ListNode(7)
head.bottom.bottom = ListNode(8)
head.bottom.bottom.bottom = ListNode(30)

head.next.bottom = ListNode(20)

head.next.next.bottom = ListNode(22)

head.next.next.next.bottom = ListNode(35)
head.next.next.next.bottom.bottom = ListNode(50)

head.next.next.next.next = ListNode(40)
head.next.next.next.next.bottom = ListNode(45)

print('Input Linked List:')
printFlattenedList(head)
head = flattenLinkedList(head)
print('\nFlattened Linked List:')
printFlattenedList(head)
````

Given a singly linked list of sizeÂ **N**. The task is toÂ **left-shift**Â the linked list byÂ **k**Â nodes, whereÂ **k**Â is a given positive integer smaller than or equal to length of the linked list.

**Example 1:**

```
Input:
N = 5
value[] = {2, 4, 7, 8, 9}
k = 3
Output:8 9 2 4 7
Explanation:Rotate 1:4 -> 7 -> 8 -> 9 -> 2
Rotate 2:Â 7Â -> 8Â -> 9Â -> 2Â -> 4
Rotate 3:Â 8Â -> 9Â -> 2Â -> 4Â -> 7

```

**Example 2:**

</aside>

````python
class ListNode:
    def __init__(self, val=0):
        self.val = val
        self.next = None


def leftShiftLinkedList(head, k):
    if head is None or head.next is None or k == 0:
        return head

    length = 0
    curr = head

    # Find the length of the linked list
    while curr:
        curr = curr.next
        length += 1

    # Adjust k if it exceeds the length of the linked list
    k = k % length

    # No shift required
    if k == 0:
        return head

    # Find the k+1th node from the beginning
    fastPtr = head
    for _ in range(k):
        fastPtr = fastPtr.next

    # Find the new head and tail
    slowPtr = head
    while fastPtr.next:
        slowPtr = slowPtr.next
        fastPtr = fastPtr.next

    # Perform left shift
    newHead = slowPtr.next
    slowPtr.next = None
    fastPtr.next = head

    return newHead


# Create a linked list from a list of values
def createLinkedList(values):
    if not values:
        return None

    head = ListNode(values[0])
    curr = head
    for val in values[1:]:
        curr.next = ListNode(val)
        curr = curr.next

    return head


# Print the values in a linked list
def printLinkedList(head):
    values = []
    curr = head
    while curr:
        values.append(curr.val)
        curr = curr.next
    print(' '.join(str(val) for val in values))


# Test the code with example inputs
values = [2, 4, 7, 8, 9]
k = 3
head = createLinkedList(values)
print('Input Linked List:')
printLinkedList(head)
head = leftShiftLinkedList(head, k)
print('\nLinked List after left shift:')
printLinkedList(head)
````

ðŸ’¡ **Question 8**

Given theÂ `head`Â of a linked list, we repeatedly delete consecutive sequences of nodes that sum toÂ `0`Â until there are no such sequences.

After doing so, return the head of the final linked list.Â  You may return any such answer.

(Note that in the examples below, all sequences are serializations ofÂ `ListNode`Â objects.)

**Example 1:**

```
Input: head = [1,2,-3,3,1]
Output: [3,1]
Note: The answer [1,2,1] would also be accepted.

```

**Example 2:**

```
Input: head = [1,2,3,-3,4]
Output: [1,2,4]

```

**Example 3:**

```
Input: head = [1,2,3,-3,-2]
Output: [1]
```

</aside>

````python
class ListNode:
    def __init__(self, val=0):
        self.val = val
        self.next = None


def removeZeroSumSublists(head):
    dummy = ListNode(0)
    dummy.next = head

    stack = []
    curr = dummy

    # Traverse the linked list
    while curr:
        # Calculate the running sum
        running_sum = curr.val
        found_zero_sum = False

        # Check if there is a sequence with zero sum
        while stack and running_sum == 0:
            running_sum -= stack.pop()
            found_zero_sum = True

        if found_zero_sum:
            curr = curr.next
            continue

        # Add the running sum to the stack
        stack.append(curr.val)
        curr = curr.next

    # Construct the final linked list
    result_head = ListNode(0)
    result_curr = result_head
    while stack:
        result_curr.next = ListNode(stack.pop(0))
        result_curr = result_curr.next

    return result_head.next


# Create a linked list from a list of values
def createLinkedList(values):
    if not values:
        return None

    head = ListNode(values[0])
    curr = head
    for val in values[1:]:
        curr.next = ListNode(val)
        curr = curr.next

    return head


# Convert a linked list to a list of values
def linkedListToValues(head):
    values = []
    curr = head
    while curr:
        values.append(curr.val)
        curr = curr.next

    return values


# Test the code with example inputs
values = [1, 2, -3, 3, 1]
head = createLinkedList(values)
print('Input Linked List:')
print(linkedListToValues(head))
head = removeZeroSumSublists(head)
print('\nLinked List after removing zero sum sublists:')
print(linkedListToValues(head))
````
